# STEP 4 — Cleaning

## Check missing values

```python
df.isnull().sum()
```

## Remove rows with missing CustomerID

```python
df = df.dropna(subset=["CustomerID"])
#Rows without a CustomerID are removed to ensure accurate customer-level analysis.
```

# STEP 6 — Customer Analysis (RFM)

## Number of Active Customers

```python
active_customers = df["CustomerID"].nunique()
active_customers
```

## Define the Reference Date
We use the last invoice date + 1 day.

```python
reference_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)
reference_date
```

## Build the RFM Table
### RFM Metrics Definition

Recency (R): Days since last purchase

Frequency (F): Number of invoices

Monetary (M): Total spending

```python
rfm = (
    df.groupby("CustomerID")
      .agg({
          "InvoiceDate": lambda x: (reference_date - x.max()).days,
          "InvoiceNo": "nunique",
          "TotalAmount": "sum"
      })
      .reset_index()
)

rfm.columns = ["CustomerID", "Recency", "Frequency", "Monetary"]
rfm.head()
```

## RFM Scoring (1_4)

We use quartiles.

```python
rfm["R_score"] = pd.qcut(
    rfm["Recency"].rank(method="first"),
    4,
    labels=[4,3,2,1]
)

rfm["F_score"] = pd.qcut(
    rfm["Frequency"].rank(method="first"),
    4,
    labels=[1,2,3,4]
)

rfm["M_score"] = pd.qcut(
    rfm["Monetary"].rank(method="first"),
    4,
    labels=[1,2,3,4]
)

rfm.head()
#Low Recency → better customer → higher score

#High Frequency & Monetary → better customer
```

## RFM Combined Score

```python
rfm["RFM_Score"] = (
    rfm["R_score"].astype(str) +
    rfm["F_score"].astype(str) +
    rfm["M_score"].astype(str)
)

rfm.head()
```

```python
# Frequency Distribution
sns.histplot(data = rfm, x = 'Frequency', bins = 30)
plt.title('Frequency Distribution')
plt.show()
```

```python
# Monetary Distribution
sns.histplot(data = rfm, x = 'Monetary', bins = 30)
plt.title('Monetary Value Distribution')
plt.show()
```

## Customer Segmentation


```python
rfm.columns
```

```python
#Mapping segments to their corresponding customer segments
segment_map = {
    r'[1-2][1-2]' : 'Hibernating',
    r'[1-2][3-4]' : 'At-Risk',
    r'[1-2]5' : "Can\'t Loose",
    r'3[1-2]' : 'About to Slip',
    r'33' : 'Need Attention',
    r'[3-4][4-5]' : 'Loyal Customers',
    r'41' : 'Promising',
    r'51' : 'New Customers',
    r'[4-5][2-3]' : 'Potential Loyalists',
    r'5[4-5]' : 'Champions',
}
#Mapping RFM segments to corresponding customer segments
rfm['RFM_Score'] = rfm['RFM_Score'].replace(segment_map, regex = True)
rfm
```

## Identification of Star Products

(High sales volume + high recurrence)

```python
product_performance = (
    df.groupby("Description")
    .agg(
        Total_Quantity=("Quantity", "sum"),
        Total_Revenue=("TotalAmount", "sum"),
        Order_Frequency=("InvoiceNo", "nunique")
    )
    .reset_index()
)

product_performance.head()
```

## Visualization


##  Top 10 Products by Revenue

```python
import matplotlib.pyplot as plt

top_products_revenue.plot(
    kind="bar",
    figsize=(10,4),
    title="Top 10 Products by Revenue"
)
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

## Identification of Dead / Low-Performing Products
Criteria (example):

Very low quantity sold

Very few orders

Low revenue contribution

```python
low_performing_products = product_performance[
    (product_performance["Total_Quantity"] <= 5) &
    (product_performance["Order_Frequency"] <= 2)
].sort_values("Total_Revenue")

low_performing_products.head(10)
```

## Contribution of Products to Total Revenue


### Percentage Contribution

```python
total_revenue = df["TotalAmount"].sum()

product_performance["Revenue_Contribution_%"] = (
    product_performance["Total_Revenue"] / total_revenue * 100
)

product_performance.sort_values(
    "Revenue_Contribution_%",
    ascending=False
).head(10)
```

### Pareto Analysis (80/20 Rule)

```python
product_performance = product_performance.sort_values(
    "Total_Revenue", ascending=False
)

product_performance["Cumulative_Revenue_%"] = (
    product_performance["Total_Revenue"].cumsum() / total_revenue * 100
)

product_performance.head(10)
```

### Visualization — Revenue Contribution

```python
import matplotlib.pyplot as plt

top_products = product_performance.head(10)

plt.figure(figsize=(10,4))
plt.bar(top_products["Description"], top_products["Revenue_Contribution_%"])
plt.xticks(rotation=75)
plt.title("Top 10 Products Contribution to Revenue (%)")
plt.ylabel("Revenue Contribution (%)")
plt.show()
```
